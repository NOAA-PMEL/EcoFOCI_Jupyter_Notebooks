## Testing Holoviews, Geoviews (and xarray)

CFSv2 Grib2 data - from https://www.ncdc.noaa.gov/data-access/model-data/model-datasets/climate-forecast-system-version2-cfsv2#CFSv2%20Operational%20Analysis

Initial attempt was to plot CFSv2 grib data.  Although, with the appropriate add on, I can read the grib file fine... I couldn't plot it.  In order to validate Geoviews works, I plotted some NARR data (as netcdf files) with no challenges.

Xarray and cartopy do a fine job plotting the data too (and natively read grib).

Reading and converting grib to netcdf data via xarray would negate the grib format being the challenge

```python
import numpy as np
import xarray as xr
import holoviews as hv
import geoviews as gv
import geoviews.feature as gf
from cartopy import crs


gv.extension('bokeh', 'matplotlib')
```

```python
(gf.ocean + gf.land + gf.ocean * gf.land * gf.coastline * gf.borders).options(
    'Feature', projection=crs.Geostationary(), global_extent=True, height=325
).cols(3)
```

```python
cfsv2_infile = '/Users/bell/Downloads/tmp2m.l.gdas.201802.grib2.nc'
nctest_infile = '/Users/bell/in_and_outbox/data_sets/reanalyis_data/NARR/3hr/air.2m.2014.nc'
```

```python
#xdf = xr.open_dataset(cfsv2_infile, engine='cfgrib')
xdf = xr.open_dataset(cfsv2_infile)
kdims = ['time', 'step', 'longitude', 'latitude']
vdims = ['t2m']
```

```python
xdf_narr = xr.open_dataset(nctest_infile, decode_coords=True)
kdims_narr = ['time', 'lon', 'lat']
vdims_narr = ['air']
```

```python
%%output backend='matplotlib', max_frames=800
%%opts QuadMesh [colorbar=True fig_size=300 projection=crs.Robinson()] (cmap='RdBu_r')
dataset = gv.Dataset(xdf_narr.sel(time=slice('2014-1-1','2014-1-2')),kdims_narr,vdims_narr)

#setting are all inline
#dataset.to(gv.QuadMesh, ['lon', 'lat']).options(cmap='viridis', colorbar=True, fig_size=200) * gf.coastline()

#settings are in the opts above
dataset.to(gv.QuadMesh, ['lon', 'lat'], dynamic=True) * gf.coastline()
```

```python
import matplotlib.pyplot as plt

air = xdf_narr['air'].isel(time=0)
ax = plt.axes(projection=crs.LambertConformal(85))
plt.pcolormesh(air.lon, air.lat, air.data, cmap='RdBu_r')
ax.coastlines()
```

```python
%%output backend='matplotlib'

dataset = gv.Dataset(xdf,vdims='t2m')
dataset.to(gv.QuadMesh, ['longitude', 'latitude'], dynamic=True)[::1].options(cmap='viridis', colorbar=True, fig_size=200) * gf.coastline()
```
