```python
infile_hgt = '/Users/bell/in_and_outbox/data_sets/reanalyis_data/NARR/daily/pressure/hgt.199410.nc'
infile_sfc = '/Users/bell/in_and_outbox/data_sets/reanalyis_data/NARR/daily/air.2m.1994.nc'
```

```python
xdf_sfc = xr.open_dataset(infile_sfc)
xdf_hgt = xr.open_dataset(infile_hgt)

kdims_sfc = ['time', 'lon', 'lat']
kdims_hgt = ['time', 'level', 'lon', 'lat']
```

```python
%%output backend='matplotlib', max_frames=800
%%opts QuadMesh [colorbar=True fig_size=450 projection=crs.Robinson()] (cmap='RdBu_r')
dataset = gv.Dataset(xdf_sfc,kdims_sfc,['air'])

#setting are all inline
#dataset.to(gv.QuadMesh, ['lon', 'lat']).options(cmap='viridis', colorbar=True, fig_size=200) * gf.coastline()

#settings are in the opts above
dataset.to(gv.QuadMesh, ['lon', 'lat'], dynamic=True) * gf.coastline()
```

```python
import matplotlib.pyplot as plt

air = xdf_sfc['air'].isel(time=0)
ax = plt.axes(projection=crs.Orthographic(-140,45))
plt.pcolormesh(air.lon, air.lat, air.data, cmap='RdBu_r', transform=crs.PlateCarree())
ax.set_global(); ax.coastlines();
```
