# Subsampling Timeseries Data in Python Using Pandas

Steps to be performed in the software
- standard python importation of a library
- defining and loading the dataset
- resampling the dataset
- saving the data out for future use

We will also look at the data as we go for good measure

```python
#import the library we need to use, in this case it is one called pandas
import pandas as pd
```

```python
filein = 'data/sd-1033_data.csv'
```

```python
#this reads in a csv, skipping the headerlines and using the time as the index parameter
# - time must be the index for the built-in sampling functions to work
df = pd.read_csv(filein,skiprows=62,parse_dates=True,index_col='TIM')
```

```python
df
```

```python
df.resample('1H').first() #aggregates all data by hour and takes the first of the batch, there are many other ways to downsample
```

```python
df.resample('6H').first() #same as above for 6-hour batches
```

```python
#save file as csv (without headers) and rename the ending to be identifiable
df.resample('6H').first().to_csv(filein.replace('.csv','_6hr.csv'))
```

*NOTICE* for each resampling we drastically cut down from 216000 samples to 600 samples... which will be much easier to collocate with the ~few hundred CTD samples
