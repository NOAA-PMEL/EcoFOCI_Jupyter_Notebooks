### Read and clean Ice Data From Strausz

```python
import pandas as pd
```

```python
datafile = '/Users/bell/in_and_outbox/2019/mordy/wisegarver/ice_conc/IceConc_C1-C3_50km.xlsx'

sheet_name='CK3 50km'
```

```python
df = pd.read_excel(datafile,sheet_name=sheet_name)
```

```python
df
```

```python
dfs = pd.melt(df,id_vars=['DOY'],value_vars=[2010,2011,2012,2013,2014,2015,2016,2017,2018,2019],
        var_name="Year", 
        value_name="IceConc")
dfs
```

```python
dfs['datetime']=pd.to_datetime(dfs['Year']*1000+dfs['DOY'],format='%Y%j')
```

```python
dfs.set_index(pd.DatetimeIndex(dfs['datetime']),inplace=True)
```

```python
dfs.resample('1H').mean().to_csv(sheet_name.replace(' ','_')+'.csv')
```
