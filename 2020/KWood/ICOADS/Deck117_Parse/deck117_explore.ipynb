# Deck 117 Explore

https://icoads.noaa.gov/deck117.html

```python
import pandas as pd
import numpy as np

import datetime
```

```python
with open('fosdic_cd117','r') as f:
    file = f.readlines()
```

```python
counter = 0
data = {}
for count, row in enumerate(file):
    if (row[:8] == '00000117'):
        print(f'Headerlines: {count}')
    else:
        try:
            datestr = '19'+row[4:6]+'-'+row[6:8]+'-'+row[8:10]+' '+row[15:16]+':00:00'
        except:
            datestr= np.nan
            
        data.update({counter:{'octant':'?',
                    'latitude':'?',
                    'longitude':'?',
                    'year':row[4:6].strip(),
                    'month':row[6:8].strip(),
                    'day':row[8:10].strip(),
                    'hour':row[15:16].strip(),
                    'shipNo1stchar':row[0],
                    'shipNo':row[1:4],
                    'datetime':datestr
                    }})
        counter +=1
```

```python
df = pd.DataFrame.from_dict(data, orient='index')
```

```python
#remove empty dates/times
df = df.drop(np.where(df['datetime'] == '19  -  -     :00:00')[0])
```

```python
df['datetime'] = pd.to_datetime(df['datetime'],errors='coerce')
```

```python
print('ShipID 3char, Obs Count, Earliest Obs Date, Latest Obs Date')
for g, data in df.groupby('shipNo'):
    if data.shipNo.count() > 2:
        print(f'{g},{data.shipNo.count()},{data.datetime.min()},{data.datetime.max()}')
```

```python
for g, data in df.groupby('shipNo1stchar'):
    print(g,data.shipNo1stchar.count())
```
