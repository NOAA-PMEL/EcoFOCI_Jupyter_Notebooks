# For Ladd GRL Paper

Ingest file from erddap and grid to depth to 1m and divenumber

```python
from erddapy import ERDDAP
import pandas as pd
import numpy as np
import xarray as xa
```

```python
server_url = 'http://pavlof.pmel.noaa.gov:8080/erddap'
```

```python
e = ERDDAP(server=server_url)
df = pd.read_csv(e.get_search_url(response='csv', search_for='sg'))
print(df['Dataset ID'].values)
```

```python
dataset_id='sg401_2017'

try:
    d = ERDDAP(server=server_url,
        protocol='tabledap',
        response='csv'
    )
    d.dataset_id = dataset_id
    d.variables = ['profileid','time','depth','latitude','longitude','wlbb2fl_sig695nm_adjusted','salinity','salinity_qc','temperature','temperature_qc']
except:
    print('Failed to generate url {}'.format(dataset_id))

try:
    df_m = d.to_pandas(
                index_col='time (UTC)',
                parse_dates=True,
                skiprows=(1,)  # units information can be dropped.
                )
    df_m.sort_index(inplace=True)
    df_m.columns = [x[1].split()[0] for x in enumerate(df_m.columns)]

except:
    print(f"something failed in data download {dataset_id}")
    pass
```

```python
for group in df_m.groupby('profileid').groups:
    print(group)
    tmpa = (df_m.groupby('profileid').get_group(group))
    tmp = tmpa.groupby(pd.cut(tmpa.depth,bins=range(0,101),labels=range(1,101))).mean()
    tmp_lat = tmpa.latitude[0]
    tmp_lon = tmpa.longitude[0]
    tmp.drop(['depth','latitude','longitude','salinity_qc','temperature_qc'],axis=1,inplace=True)
    tmp['datetime'] = tmpa.index[0]
    tmp['latitude'] = tmp_lat
    tmp['longitude'] = tmp_lon
    tmp.to_csv(group+'.csv')
```
