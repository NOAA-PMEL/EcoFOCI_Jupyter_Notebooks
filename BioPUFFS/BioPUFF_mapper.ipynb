```python
weeklygroupbed = df_merged[df_merged.year==2021].groupby('weeks')
```

## Make plots

```python
#using xarray for bathymetry data read
import xarray as xa

extent=[-179.5, -145, 50, 74]

server_url = 'https://coastwatch.pfeg.noaa.gov/erddap/griddap/usgsCeSS111'
bathy = xa.open_dataset(server_url) 
bathy_sub = bathy.sel(latitude=slice(extent[3],extent[2]),longitude=slice(extent[0],extent[1])).load()
```

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import cmocean
from cartopy.mpl.ticker import LongitudeFormatter, LatitudeFormatter

def make_map(projection=ccrs.PlateCarree()):
    fig, ax = plt.subplots(figsize=(12,12),
                           subplot_kw=dict(projection=projection))
    if projection == ccrs.PlateCarree():
        gl = ax.gridlines(draw_labels=True)
        gl.xlabels_top = gl.ylabels_right = False
        gl.xformatter = LONGITUDE_FORMATTER
        gl.yformatter = LATITUDE_FORMATTER
    return fig, ax
```

```python
#this is fast if topo is turned off
with_topo = False

projection=ccrs.LambertConformal(central_longitude=-160.0)
transformation=ccrs.PlateCarree()

land_50m = cfeature.NaturalEarthFeature('physical', 'land', '50m',
                                                edgecolor='face',
                                                facecolor='1.0')



for wn, dftemp in weeklygroupbed:
    fig,ax = make_map(projection=projection)

    #plt.colorbar(c)
    ax.add_feature(land_50m)
    ax.coastlines(resolution='10m')
    ax.set_extent(extent)

    if with_topo:
        ax.pcolormesh(bathy_sub.longitude, bathy_sub.latitude, 
                    bathy_sub.topo, cmap=cmocean.cm.gray,
                    vmin=-3000,vmax=100,
                    transform=transformation)

    c = ax.scatter(dftemp.longitude, dftemp.latitude, s=100, c=dftemp.sr_temp, 
                   cmap=cmocean.cm.balance, vmin=-2, label=('2021: Week '+str(wn)),
                               transform=transformation)
    cbar = fig.colorbar(c, orientation='vertical', extend='both',shrink=0.75)
    cbar.ax.set_ylabel('Temperature ($^\circ$C)')
    
    l = plt.legend(loc='lower left', bbox_to_anchor=(0.01, 0.01), fontsize='xx-large', fancybox=True, frameon=False, markerscale=0)
    
    for text in l.get_texts():
        text.set_color("white")
    
    # fig.savefig('Bio_PUFF_2021_week'+str(wn).zfill(2)+'.png',dpi=300)

        
```
