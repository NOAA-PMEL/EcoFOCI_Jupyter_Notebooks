```python
import matplotlib.pyplot as plt

import cartopy
import cartopy.crs as ccrs
from cartopy.io import shapereader
from cartopy.mpl.gridliner import LONGITUDE_FORMATTER, LATITUDE_FORMATTER


def make_map(projection=ccrs.PlateCarree()):
    fig, ax = plt.subplots(figsize=(9, 13),
                           subplot_kw=dict(projection=projection))
    gl = ax.gridlines()
    if (projection==ccrs.PlateCarree()) or (projection==ccrs.Mercator()):
        gl.xlabels_top = gl.ylabels_right = False
        gl.xformatter = LONGITUDE_FORMATTER
        gl.yformatter = LATITUDE_FORMATTER
    return fig, ax

```

```python
from cartopy.feature import NaturalEarthFeature

coast = NaturalEarthFeature(category='physical', scale='110m',
                            facecolor='black', name='coastline')

fig, ax = make_map(projection=ccrs.Orthographic(-135, 90))


#feature = ax.add_feature(coast, edgecolor='black')
feature = ax.add_feature(cartopy.feature.LAND, zorder=0, edgecolor='black', facecolor='grey')
```

```python
fig.savefig('figures/NP_Ortho_Greyland_coarseboundaries.png',dpi=300)
fig.savefig('figures/NP_Ortho_Greyland_coarseboundaries_transparent.png', transparent=True,dpi=300)
```
