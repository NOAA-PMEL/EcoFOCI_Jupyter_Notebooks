```python
import pygmt as gmt
```

```python
fig = gmt.Figure()


fig.coast(region=[-180, -140, 50, 80], projection='B-160/60/55/60/6i', land='black', water='lightgrey',
          frame=True,U=True) 
```

```python
fig.show()
```

```python
fig = gmt.Figure()


fig.coast(region=[-180, -140, 50, 80], projection='B-160/60/55/60/6i', land='blanchedalmond', water='lightblue',
          frame=True,U=True) 

fig.show()
```

```python
# Load the builtin Earth relief grid as an xarray.DataArray.
relief = gmt.datasets.load_earth_relief(resolution="10m")

fig = gmt.Figure()


fig.basemap(region=[-180, -140, 50, 80], projection='B-160/60/55/60/6i',frame=True) 
fig.grdimage(relief, cmap="bathy")
fig.coast(land='black')
fig.show()
```
