## Mapping of EcoFOCI Glider Locations from Science Data Set - small selection

**Dives 0-999** are shown here but focus on a smaller subset for analysis

```python
import gmt
import xarray as xa
```

```python
data_dir = '/Users/bell/Programs/Python/EcoFOCI_OculusGlider/data/one_dtdz/'
```

```python
xaf = xa.open_mfdataset(data_dir+'p4010*_m.nc',autoclose=True)
xad = xa.open_mfdataset(data_dir+'p4010*_d.nc',autoclose=True)
xau = xa.open_mfdataset(data_dir+'p4010*_u.nc',autoclose=True)
```

```python
sfc_lat = xaf.latitude[xaf.Pressure==1]
sfc_lon = xaf.longitude[xaf.Pressure==1]

#print(max(sfc_lon),max(sfc_lat))
```

```python
fig = gmt.Figure()


fig.coast(region=[-175, -170, 59, 61], projection='B-170/60/55/60/6i', land='black', water='lightgrey',
          frame=True,U=True) 
```

```python
fig.show()
```

Below is a plot of the sfc latitude/longitude.  Blue points met the criterion to have the salinity corrected, red did not (or green if upcasts are being plotted)

```python
fig.plot(x=sfc_lon, y=sfc_lat, color='blue', style='c0.1c')
fig.plot(x=xad.longitude[xad.Pressure==1], y=xad.latitude[xad.Pressure==1], color='green', style='c0.1c')
fig.plot(x=xau.longitude[xau.Pressure==1], y=xau.latitude[xau.Pressure==1], color='red', style='c0.1c')
```

```python
fig.show()
```
