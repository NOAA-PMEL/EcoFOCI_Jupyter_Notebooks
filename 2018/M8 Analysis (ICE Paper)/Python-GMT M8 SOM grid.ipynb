### Load gridpoints for SOM data analysis
These are NCEP-Reanalysis locations and Satellite Grid Points

```python
import pandas as pd
```

```python
filename = 'data/coords_subset_4pt_noland.asc2'
dfi = pd.read_csv(filename,sep='\s+')

filename = 'data/coords_ncep.asc2'
dfn = pd.read_csv(filename,sep='\s+')
```

```python
import gmt
```

```python
fig = gmt.Figure()

#projection - 
fig.coast(region=[-185, -150, 55, 75], projection='B-167.5/60/55/60/6i', land='blanchedalmond', water='lightblue',
          frame=True,U=True) 
```

```python
fig.show()
```

```python
fig.plot(x=dfi.lon.values, y=dfi.lat.values, color='black', style='c0.1c')
fig.plot(x=dfn.lon.values, y=dfn.lat.values, color='red', style='c0.1c',frame='g5')
```

```python
fig.show()
```
