```python
import xarray as xa
import pandas as pd
```

```python
fp='/Users/bell/Desktop/glider404/PugetSound_24OCT2018/p4040001.nc'
```

```python
xdf = xa.open_dataset(fp)
```

```python
xdf.info
```

```python
pd.concat([xdf['wlbb2fl_time'].to_dataframe(),
           xdf['wlbb2fl_temp'].to_dataframe(),
           xdf['wlbb2fl_FL1sig'].to_dataframe(),
           xdf['wlbb2fl_700sig'].to_dataframe(),
           xdf['wlbb2fl_470sig'].to_dataframe()], axis=1).to_csv(fp.replace('.nc','_wl.csv'))
```

```python
pd.concat([xdf['auxCompass_time'].to_dataframe(),
           xdf['auxCompass_rol'].to_dataframe(),
           xdf['auxCompass_pit'].to_dataframe(),
           xdf['auxCompass_hdg'].to_dataframe()], axis=1).to_csv(fp.replace('.nc','_compass.csv'))
```

```python
pd.concat([xdf['depth_time'].to_dataframe(),
           xdf['depth_depth'].to_dataframe()], axis=1).to_csv(fp.replace('.nc','_depth.csv'))
```
