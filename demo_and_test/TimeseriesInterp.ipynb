### Interpolate timeseries to arbitrary time intervals

```python
import pandas as pd
import numpy as np

import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator, WeekdayLocator, MonthLocator, DayLocator, HourLocator, DateFormatter
import matplotlib.ticker as ticker
```

```python
fn='/Users/bell/ecoraid/2018/Moorings/18mt3a/working/18mt3a_an9_0050m.unqcd.csv'
draw =pd.read_csv(fn, parse_dates=['time'], index_col='time')
draw[:5]
```

```python
draw.resample('10T').mean().interpolate(method='time').resample('10T').median()
```

```python
draw.resample('10T').mean().interpolate(method='time').resample('10T').median().to_csv(fn.replace('.csv','.10mg.csv'))
```

```python
### truncate time
draw.resample('1h').mean().to_csv(fn.replace('.csv','.1hrg.csv'))
```
