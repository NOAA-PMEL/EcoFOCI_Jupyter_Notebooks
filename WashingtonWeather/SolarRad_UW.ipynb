## Sunny in January

I used to always visit the Seattle region in January (because MLK weekend had cheaper flights than Christmas/New Years) and I was astounded that it always seemed like a good sunny time to visit... a glimmer of beautiful weather in an otherwise gloomy location... was this just perception?

```python
%matplotlib inline
```

```python
import pandas as pd

import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/UW_radData.csv',skiprows=1,parse_dates={'datetime': ['Date', 'Time']},index_col='datetime')
```

```python
df.Solar = pd.to_numeric(df.Solar, errors='coerce')
```

```python
df.info()
```

```python
df_daily = df.resample('1D').mean()
```

```python
for year in range(2000,2019,1):
    plt.plot(df_daily[str(year)+'-01'].index.dayofyear,df_daily[str(year)+'-01']['Solar'])
```

```python
df_climo =df.groupby(df.index.dayofyear).mean()
```

```python
df_climo[0:60].Solar.plot()
```
