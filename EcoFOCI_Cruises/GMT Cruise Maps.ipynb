## Plot Cruise Maps

- python>=3.8
- using gmt (basemap is deprecated)

```python
import pygmt
import cmocean
import pandas as pd
```

```python
path = 'data/'
ops = pd.read_excel(path + 'Table 1_2020 cruise track aug13.xlsx')
```

```python
extent = [-178, -150, 52, 75]
#if extent is -179 or -180 this will fail
#if resolution is 05m, this will segfault?
grid = pygmt.datasets.load_earth_relief(region=extent, resolution='04m')

fig = pygmt.Figure()
fig.basemap(region=extent, projection='B-165/65/55/65/6i', frame=True)
fig.grdimage(grid=grid, cmap="gray")
fig.coast(shorelines=True, land='tan')
fig.plot(ops.Long, ops.Lat, pen="1.5p,green")
fig.plot(ops.Long, ops.Lat, color='black', style='s.125c')

fig.show()
```
