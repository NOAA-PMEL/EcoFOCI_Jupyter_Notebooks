#### Ingest SPN1 data
(slow to load and I think it is due to setting the index to the Time)

```python
### load spn1 data
f1 = 'data/SPN1_20170518/A1595_Mask.txt'
df1 = pd.read_csv(f1,header=4,delimiter='\t',parse_dates=[0],low_memory=False)
df1 = df1.set_index(pd.DatetimeIndex(df1['Time']))
df1.Total = pd.to_numeric(df1.Total, errors='coerce')
df1_min = df1.resample('T').mean()

f2 = 'data/SPN1_20170518/A1596_NOMASK.txt'
df2 = pd.read_csv(f2,header=3,delimiter='\t',parse_dates=[0])
df2 = df2.set_index(pd.DatetimeIndex(df2['Time']))
df2_min = df2.resample('T').mean()
```

#### Ingest flex data

```python
### load flex data
f3 = 'data/SPN1_20170518/flex0010-native.swr'
df3 = pd.read_csv(f3,header=5,delimiter='\s+', 
		names=['Date','Time', 'Total', 'StdDev', 'Nsec'], 
		na_values = [1E+35], parse_dates=[['Date', 'Time']])
df3 = df3.set_index(pd.DatetimeIndex(df3['Date_Time']))

f4 = 'data/SPN1_20170518/flex1000-native.swr' # truth?
df4 = pd.read_csv(f4,header=5,delimiter='\s+', 
		names=['Date','Time', 'Total', 'StdDev', 'Nsec'], 
		na_values = [1E+35], parse_dates=[['Date', 'Time']])
df4 = df4.set_index(pd.DatetimeIndex(df4['Date_Time']))
```

#### Add the nearby UW rooftop shortwave sensor
[UW rooftop data site](http://www-k12.atmos.washington.edu/k12/grayskies/plot_nw_wx.cgi?Measurement=Solar&station=UWA&interval=168&timezone=7&rightlab=y&connect=lines&groupby=overlay&begmonth=1&begday=1&begyear=2017&beghour=0&endmonth=1&endday=1&endyear=2017&endhour=0)

```python
# UW data from rooftop
f5 = 'data/SPN1_20170518/UW_Radn_rooftop.xlsx'
df5 = pd.read_excel(f5,parse_dates=[['Date', 'Time']])
df5 = df5.set_index(pd.DatetimeIndex(df5['Date_Time']))
```

```python
plt.figure()
plt.subplot(1,1,1)
df1_min.plot(x=df1_min.index,y=['Total'], ax=plt.gca())
df2_min.plot(x=df2_min.index,y=['Total'], ax=plt.gca())
df2_min.plot(x=df2_min.index,y=['Diffuse'], ax=plt.gca())
df3.plot(x=df3.index,y=['Total'], ax=plt.gca())
df4.plot(x=df4.index,y=['Total'], ax=plt.gca())
df5.plot(x=df5.index,y=['Radn'], ax=plt.gca())
ax1=plt.gca()
ax1.legend(['A1595','A1596 Center', 'A1596','flex 0010','flex 1000','UW rooftop'])
```

The periods where the curves transition smoothly through out the day (and wich represent the theoretical daily incoming solar radiation function) are the periods we are most interested in for intercomparisons.  We want to minimize the number of features that are playing a role in the measurement, so clouds, shadows, haze and the like are periods that provide more challenging comparisons.

The evening of the 21st through the morning of the 22nd are ideal as is the 23rd (data from UW but not yet available for the SPN1)

### Lets look at the difference between instruments if we select the flex1000 as our "truth" instrument

```python
plt.figure(figsize=(16,8))
plt.subplot(1,1,1)
plt.plot(df4_sub.index,df4_sub['Total']-df1_sub['Total'],'.',markersize=5)
plt.plot(df4_sub.index,df4_sub['Total']-df2_sub['Total'],'.',markersize=5)
plt.plot(df4_sub.index,df4_sub['Total']-df2_sub['Diffuse'],'.',markersize=5)
plt.plot(df4_sub.index,df4_sub['Total']-df3_sub['Total'],'.',markersize=5)
ax1=plt.gca()
ax1.legend(['flex1000-A1595','flex1000-A1596 Center', 'flex1000-A1596','flex1000-flex0010'])
ax1.set_ylim([-200,200])
```

Differences that appear to be larger during midday (when the solar signal is larger) and smaller in the evenin/morning hours may be due to instruments that are out of calibration or may indicate characteristics unique to one sensor (dirt on dome, shadows, asymetries, etc.)

Uniform differences in calibration or characterization can often be corrected by applying a scaling factor to the instrument in question in order to make it match the reference instrument.  Plotting the data of one instrument against the data of the reference instrument should give a 1:1 relationship if they were exactly the same.  If they have equivalent responses (but just of different magnitudes) or there is a non-zero baseline offset, 1-to-1 plots will also show this.  

Below we see that A1595 and the flex1000 (topleft) correspond well  (although the A1595 SPN1 is reading high and can be corrected by the slope of the regression line)  A1596 Center Detector only also corresponds eell with the flex1000 but the averaged other detectors reports consistently lower values (but probably not out of cal expectations).  The two flex systems have an obvious scaling difference. *Communication with Jennifer Keane suggests the flex1000 has an older calibration than flex0010 - **this needs evaluation to know which is "truth"** *

One questions is that in both the tflex0010 and A1596 Instruments, there are two distinct clusters of points along the 1:1 line.  I am assuming that this indicative of differing morning and afternoon response characteristics... aka the sensors are not truly independent of direction/orientation.
